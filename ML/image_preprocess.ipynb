```python
import torch
import torch.nn as nn
import torchvision
import torchsummary
from PIL import Image
import os
from concurrent.futures import ProcessPoolExecutor
import concurrent.futures
from tqdm.auto import tqdm

print(torch.cuda.get_device_name())
```

```python
def crop_resize(source_path, output_path) -> tuple[str, int]:
    processed_images = 0
    os.makedirs(output_path, exist_ok = True)
    for image_name in tqdm(os.listdir(source_path), leave=False):
        output_image_path = os.path.join(output_path, image_name)
        if os.path.exists(output_image_path):
            processed_images += 1
            continue
        if not "jpg".lower() in image_name.lower():
            continue
        try:
            image = Image.open(os.path.join(source_path, image_name))
            new_width = 1200
            new_height = 1200
            width, height = image.size
            left = (width - new_width)/2
            top = (height - new_height)/2
            right = (width + new_width)/2
            bottom = (height + new_height)/2
            image = image.crop((left, top, right, bottom))
            new_size = (224, 224)
            image = image.resize(new_size)
            image.save(os.path.join(output_path, image_name))
            processed_images += 1
        except Exception as e:
            print(f'{source_path}: {e}', flush=True)
    return (source_path, output_path, processed_images)
```

```python
# 안되네요?
with ProcessPoolExecutor(max_workers=8) as executor:
    # image preprocessing
    source_root = 'e:\\pill_image'
    target_root = 'e:\\pill_image_cropped'
    args = []
    for item_seq in tqdm(os.listdir(source_root)):
        image_root = os.path.join(source_root, item_seq)
        if not os.path.isdir(image_root):
            continue
        image_names = os.listdir(image_root)
        output_root = os.path.join(target_root, item_seq)
#         futures.append(executor.submit(crop_resize, image_root, output_root))
        args.append((image_root, output_root))
    for result in tqdm(executor.map(crop_resize, *zip(*args)), total=len(args)):
        print(f'{result[0]} -> {result[1]}: Processed {result[2]} images')
```
